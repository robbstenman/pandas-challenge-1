## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
display(df[['unit_price', 'qty', 'line_subtotal']].head())
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
df['total_weight'] = df['unit_weight'] * df['qty'] 
df['shipping_price'] = df.apply(lambda row: row['total_weight'] * 7 
    if row['total_weight'] > 50
    else row['total_weight'] * 10, axis=1)
display(df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3))
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['pretax_subtotal'] = df['line_subtotal'] + df['shipping_price']
df['line_price'] = df['pretax_subtotal'] * (1 +.0925)
df_display = (df[['line_subtotal', 'shipping_price', 'line_price']].head(3)).style.format({
'line_subtotal': '{:.2f}',
'shipping_price': '{:.2f}',
'line_price': '{:.2f}'
})
df_display
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
format_dict = {col: "{:.2f}" for col in df.select_dtypes(include=['float64', 'int64']).columns}
df_styled = df.head(3).style.format(format_dict)
df_styled
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['line_profit'] = df['line_price'] - df['line_cost']
format_dict = {col: "{:.2f}" for col in df.select_dtypes(include=['float64', 'int64']).columns}
df_styled = df.head(3).style.format(format_dict)
df_styled
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top_five = [33615, 66037, 46820, 38378, 24741]
top_five_df = df[df['client_id'].isin(top_five)]
total_spent = top_five_df.groupby('client_id')['line_price'].sum().reset_index()
total_spent.columns = ['client_id', 'top_five_amount']
total_spent['top_five_amount'] = total_spent['top_five_amount'].apply(lambda x: f"${x:,.2f}")

display(total_spent)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
top_five = [33615, 66037, 46820, 38378, 24741]
top_five_df = df[df['client_id'].isin(top_five)]

client_total_data = []

for client_id in top_five:
    client_df = top_five_df[top_five_df['client_id'] == client_id]
    
    client_total_data.append({
        'client_id': client_id,
        'qty': client_df['qty'].sum(),
        'shipping_price': round(client_df['shipping_price'].sum(), 2),
        'line_price': round(client_df['line_price'].sum(), 2),
        'line_cost': round(client_df['line_cost'].sum(), 2),
        'line_profit': round(client_df['line_profit'].sum(), 2)
    })    
    
summary_df = pd.DataFrame(client_total_data)
    
summary_df['client_id'] = summary_df['client_id'].astype(int)

sorted_summary_df = summary_df.sort_values('line_profit', ascending=False)

print(sorted_summary_df)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
summary_df = summary_df.rename(columns={
    'client_id': 'Client ID',
    'qty': 'Quantity',
    'shipping_price': 'Total Shipping Price (M$)',
    'line_price': 'Total Price (M$)',
    'line_cost': 'Total Cost (M$)',
    'line_profit': 'Total Profit (M$)'
})

for column in ['Total Shipping Price (M$)', 'Total Price (M$)', 'Total Cost (M$)', 'Total Profit (M$)']:
    summary_df[column] = summary_df[column] / 1_000_000
    
for column in ['Total Shipping Price (M$)', 'Total Price (M$)', 'Total Cost (M$)', 'Total Profit (M$)']:
    summary_df[column] = summary_df[column].apply(lambda x: f"${x:,.2f}")
    
sorted_summary_df = summary_df.sort_values('Total Profit (M$)', ascending=False)

display(sorted_summary_df)
```
